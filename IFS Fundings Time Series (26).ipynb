```python
import pandas as pd
import numpy as np
import csv
import datetime as dt
from datetime import timedelta
import time
import matplotlib as plt
import statistics
import math
```

```python
adr = pd.read_csv('analytics_data_report')
df = pd.DataFrame(adr)
df.columns = df.columns.str.replace(' ', '_').str.lower()
bookcount = df.groupby("partner")["loan_booked"].count()
leadcount = df.groupby("partner")["referral_received"].count()
conversions = bookcount / leadcount
```

```python
t1 = pd.to_datetime(df['referral_received'])
t2 = pd.to_datetime(df['loan_booked'])
df['difference'] = t2 - t1
df['difference'] = df['difference'] / dt.timedelta(days=1)
df_dates = df[['partner', 'referral_received', 'loan_booked', 'difference']]
mean_diff = df_dates.groupby("partner")["difference"].mean()
med_diff = df_dates.groupby("partner")["difference"].median()
df_summ = pd.DataFrame({'Volume': leadcount,
                       'Conversions': conversions,
                       'Mean_diff': mean_diff,
                       'Med_diff': med_diff})
df_summ.dropna()
```
